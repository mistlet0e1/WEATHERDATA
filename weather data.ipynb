```python
import pandas as pd
import numpy as np

# Read the Excel file
df = pd.read_excel("Weather_Data.xlsx")

# Display first 5 rows
print(df.head())
```

```python
# Unique Wind Speeds
print(df["Wind Speed_km/h"].unique())
```

```python
# Count of Clear Weather
print(df[df["Weather"] == "Clear"].shape[0])
```

```python
# Wind Speed = 4
print(df[df["Wind Speed_km/h"] == 4])
```

```python
# Mean Visibility
print(df["Visibility_km"].mean())
```

```python
# Standard Deviation of Pressure
print(df["Press_kPa"].std())
```

```python
# Records containing Snow
print(df[df["Weather"].str.contains("Snow", case=False)])
```

```python
# Weather = Fog
print(df[df["Weather"] == "Fog"])
```

```python
# Mean of each numeric column grouped by Weather
print(df.groupby("Weather").mean(numeric_only=True))
```

```python
# Weather = Clear OR Visibility > 40
print(df[(df["Weather"] == "Clear") | (df["Visibility_km"] > 40)].shape[0])
```

```python
print(df[(df["Weather"] == "Clear") & (df["Rel Hum_%"] > 50)].shape[0])
```

```python
import matplotlib.pyplot as plt

df = pd.read_csv("Weather_Data.csv")

weather_count = df["Weather"].value_counts()

plt.figure(figsize=(12,6))
weather_count.plot(kind="bar")
plt.title("Weather Condition Frequency")
plt.xlabel("Weather Condition")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df["Temp_C"], bins=20)
plt.title("Temperature Distribution")
plt.xlabel("Temperature (°C)")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(12,5))
plt.plot(df["Temp_C"])
plt.title("Temperature Trend")
plt.xlabel("Record Number")
plt.ylabel("Temperature (°C)")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df["Temp_C"], df["Rel Hum_%"])
plt.title("Temperature vs Relative Humidity")
plt.xlabel("Temperature (°C)")
plt.ylabel("Relative Humidity (%)")
plt.show()
```

```python
weather = df["Weather"].value_counts().head(10)

plt.figure(figsize=(8,8))
plt.pie(weather, labels=weather.index, autopct="%1.1f%%")
plt.title("Top 10 Weather Conditions")
plt.show()
```

```python
avg_temp = df.groupby("Weather")["Temp_C"].mean().sort_values()

plt.figure(figsize=(12,6))
avg_temp.plot(kind="bar")
plt.title("Average Temperature by Weather")
plt.xlabel("Weather")
plt.ylabel("Average Temperature")
plt.xticks(rotation=90)
plt.show()
```

```python
#Shows the spread of wind speed values and outliers.

plt.figure(figsize=(6,5))
plt.boxplot(df["Wind Speed_km/h"])
plt.title("Wind Speed Distribution")
plt.ylabel("Wind Speed (km/h)")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))

sns.heatmap(
    df[["Temp_C","Dew Point Temp_C","Rel Hum_%","Wind Speed_km/h","Visibility_km","Press_kPa"]].corr(),
    annot=True,
    cmap="coolwarm"
)

plt.title("Correlation Heatmap")
plt.show()
```
